```python
import numpy as np
import scipy.stats as sts
import matplotlib.pyplot as plt

# Normal
plt.figure(None, (8, 4))
x = np.linspace(-5, 7, 300)
plt.plot(x, sts.norm.pdf(x, 2, 0.75), label='mu=2, sigma=0.75')
plt.plot(x, sts.norm.pdf(x, -1, 0.5), label='mu=-1, sigma=0.5')
plt.hist(np.random.normal(2, 0.75, 1000), label='histogram', density = True)
plt.legend()
plt.show()
```

```python
# Gamma
plt.figure(None, (8, 4))
x = np.linspace(-2, 10, 300)
plt.plot(x, sts.gamma.pdf(x, 2, 0.75), label='loc=2, scale=0.75')
plt.plot(x, sts.gamma.pdf(x, 3, 1.5), label='loc=3, scale=1.5')
plt.hist(np.random.gamma(2, 0.75, 1000), label='histogram', density = True)
plt.legend()
plt.show()
```

```python
# Beta
plt.figure(None, (8, 4))
x = np.linspace(-1, 2, 300)
plt.plot(x, sts.beta.pdf(x, 2, 0.75), label='shape=2, scale=0.75')
plt.plot(x, sts.beta.pdf(x, 3, 1.5), label='shape=3, scale=1.5')
plt.hist(np.random.beta(2, 0.75, 1000), label='histogram', density = True)
plt.legend()
plt.show()
```
